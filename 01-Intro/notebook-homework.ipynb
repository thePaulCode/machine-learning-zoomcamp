```python
#Vector - vector multiplication 
u = np.array([2,4,5,6])
ushape = u.shape
ushape
```

```python
v = np.array([1,0,0,2])
vshape = v.shape
vshape
```

```python
def vector_vector_multiplication(u,v):
    assert u.shape[0] == v.shape[0]

    n = u.shape[0]

    result = 0

    for i in range(n):
        result = result + u[i]*v[i]

    return result    


```

```python
vector_vector_multiplication(u,v)
```
