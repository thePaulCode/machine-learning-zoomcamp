# 2.5 Linear Regression

```python
df_train.iloc[10]
```

```python
xi = [453, 11, 86]
w0 = 7.17
w = [0.01,0.04,0.002]
```

```python
def linear_regression(xi):
    pred = w0
    n = len(xi)

    for j in range(n):
        pred = pred + xi[j]*w[j]

    return pred    
```

```python
# values in ln
linear_regression(xi)
```

```python
# real price exp^value-1
np.expm1(# values in ln
linear_regression(xi))
```
