```python
xi = [453, 11, 86]
w0 = 7.17
w = [0.01,0.04,0.002]
```

```python
def linear_regression(xi):
    pred = w0
    n = len(xi)

    for j in range(n):
        pred = pred + xi[j]*w[j]

    return pred    
```

# 2.6 Linear Regression Vector Form

```python
def dot(xi, w):
    n = len(xi)

    result = 0.0

    for j in range(n):
        result = result + xi[j] * w[j]

    return result    
```

```python
def linear_regression(xi):
    return w0 + dot(xi, w)
```

```python
linear_regression(xi)
```

```python
w_new = [w0] + w
w_new
```

```python
def linear_regression(xi):
    xi = [1] + xi
    return dot(xi, w_new)
```

```python
# How do we set w_new values?

def train_linear_regression(X, y):
    ones = np.ones(X.shape[0])
    X = np.column_stack([ones, X])

    XTX = X.T.dot(X)
    XTX_inv = np.linalg.inv(XTX)
    w_full = XTX_inv.dot(X.T).dot(y)

    return w_full[0], w_full[1:]
    
```

```python
train_linear_regression(X, y)
```
